# Q:PREDICT THE BURNED AREA OF FOREST FIRES WITH NEURAL NETWORKS


```python
import pandas as pd
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("E:\\Assisgnments\\NeuralNetwork\\forestfires.csv")
```

```python
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data=data.drop(['month','day'],axis=1)
```

```python
data
```

```python
mapping = {'small': 1, 'large': 2}
```

```python
data = data.replace(mapping)
```

```python
data
```

```python
# spliting data into x and y
x = np.array(data.iloc[:,0:28])
y = np.array(data.iloc[:,28])
y
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_norm = norm_func(x)
x_norm
```

```python
# Data spliting 
x_train,x_test,y_train,y_test= train_test_split(x_norm,y, test_size=0.2)
```

```python
# Create model
model = Sequential()
model.add(Dense(8, input_dim=28, activation='linear'))
model.add(Dense(4,  activation='tanh'))
model.add(Dense(1,  activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the model
model.fit(x_train, y_train, validation_split=0.3, epochs=120, batch_size=10)
```

```python
scores = model.evaluate(x_train, y_train)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))#train data
```

```python
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))#for test
```
